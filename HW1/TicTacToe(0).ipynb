```
from random import  randint
import random
def initialize():
  board = dict()
  for i in range(1,10):
    board[i] = i
  return board
def print_board(board):
  print('{b1}\t{b2}\t{b3}'.format(b1=board[1], b2=board[2], b3=board[3]))
  print('{b4}\t{b5}\t{b6}'.format(b4=board[4], b5=board[5], b6=board[6]))
  print('{b7}\t{b8}\t{b9}'.format(b7=board[7], b8=board[8], b9=board[9]))
  print()
def tic_tac_toe(turn):
  board = initialize()
  print_board(board)
  win_situation = [[1,2,3],[1,4,7],[1,5,9],[2,5,8],[3,5,7],[3,6,9],[4,5,6],[7,8,9]]
  
  empty_places = [i for i in range(1,10)]
  if turn == 0:
    start = randint(1,9)
    board[start] = 'x'
    empty_places.remove(start)
    for i in range(len(win_situation)):
      for j in range(len(win_situation[i])):
          if  win_situation[i][j]== start:
            win_situation[i][j] = 'x'
    
  print_board(board)
  print(win_situation)
  win = False
  while empty_places != []:
    print()
    print('choose where you put o (between(1,9) in non taken places){empty_cells}'.format(empty_cells = empty_places))
    rival_choise = int(input())
    
    empty_places.remove(rival_choise)
    
    board[rival_choise] = 'o'
    for i in range(len(win_situation)):
      for j in range(len(win_situation[i])):
        if  win_situation[i][j]== rival_choise:
          win_situation[i][j] = 'o'
        
    for i in range(len(win_situation)):
      if win_situation[i].count('o') == 3:
        win = True
        print_board(board)
        print()
        print('You won')
        return win
    for i in range(len(win_situation)):
      
      if win_situation[i].count('x') == 2 and win_situation[i].count('o') == 0:
        
        for j in range(len(win_situation[i])):
          if isinstance(win_situation[i][j], int):
            board[win_situation[i][j]] = 'x'
            empty_places.remove(win_situation[i][j])
            win_situation[i][j] = 'x'
            
            win = False
            print_board(board)
            print()
            print('You lost')
            return win
    selected = False
    choise = 0
    for i in range(len(win_situation)):
      
      if win_situation[i].count('x') == 0 and win_situation[i].count('o') == 2:
    
        for j in range(len(win_situation[i])):
          if isinstance(win_situation[i][j],int):
           
            choise = win_situation[i][j]
            
            selected = True
            break
        break
    
    if selected == False:
      
      for i in range(len(win_situation)):
        
        if win_situation[i].count('x') == 1 and win_situation[i].count('o') == 0:
          
          for j in range(len(win_situation[i])):
            if isinstance(win_situation[i][j],int):
              
              
              choise = win_situation[i][j]
              
              selected = True
              break
          break
      
    if selected == False: 
      for i in range(len(win_situation)):
        if win_situation[i].count('x') == 0 and win_situation[i].count('o') == 0:
          
          choise = random.choice(win_situation[i])
        

          selected = True
          break
          
    
            
    if selected == False:
      if empty_places != []:
        choise = random.choice(empty_places)
        
        selected = True
            
    if selected == True:
      board[choise] = 'x'
      empty_places.remove(choise)
      for i in range(len(win_situation)):
          for j in range(len(win_situation[i])):
            if  win_situation[i][j]== choise:
              win_situation[i][j] = 'x'    
               

    
    print(empty_places)
    print(win_situation)
    print_board(board)
  print()
  print('Game ended. nobody won!')
```
